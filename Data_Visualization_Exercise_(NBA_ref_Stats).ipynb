```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
nba_ref = pd.read_csv('/content/NBA Referee Stats.csv')
nba_ref.head()

```

```python
sns.set()
nba_corr = nba_ref.corr()
plt.figure(figsize=(20,20))
sns.heatmap(nba_corr);
```

```python
nba_ref.describe()
```

```python
nba_ref.shape
```

```python
sns.set()
total_fouls = nba_ref.groupby('Referee'),['Total fouls']
total_personal = nba_ref.groupby('Referee')['Personal']
total_shooting = nba_ref.groupby('Referee')['Shooting'].mean()
#Count shooting fouls reals from each Ref
total_fouls = nba_ref['Referee'].value_counts()
#Plot the counts of each cereal brand
plt.figure(figsize=(25, 10))
total_fouls.plot(kind = 'bar',)
plt.title('Number of Shooting fouls by ref', fontsize = 10)
plt.xlabel('Referee', fontsize = 20)
plt.ylabel('Shooting Fouls', fontsize = 20)
#Show your figure
plt.show();
```
